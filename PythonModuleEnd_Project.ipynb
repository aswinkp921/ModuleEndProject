## **Find out under which team and position, spending in terms of salary is high.**

```python
Team_salary = df1.groupby(["Team","Position"])["Salary"].sum()
print(f"Total Salary for Each Team and Position:\n{Team_salary}")
```

```python
sorted_team_salary = Team_salary.sort_values(ascending=False)
top_teams_positions = sorted_team_salary.head(5)
print(f"Top 5 Teams and Positions with Highest Total Salary:\n{top_teams_positions}")
```

```python
highest = Team_salary.max()
highest_spending = Team_salary[Team_salary == highest]
print(f"\nTeam and position,spending in terms of salary is high:\n{highest_spending}")
```

## **Find if there is any correlation between age and salary , represent it visually.**

```python
sns.scatterplot(x="Age", y="Salary", data=df1)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Correlation between age and salary")
plt.grid()
plt.show()
```
